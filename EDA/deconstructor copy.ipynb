```python
import os

def get_directory_size(path):
    total_size = 0
    for dirpath, dirnames, filenames in os.walk(path):
        for filename in filenames:
            filepath = os.path.join(dirpath, filename)
            if not os.path.islink(filepath):  # skip symbolic links
                total_size += os.path.getsize(filepath)
    return total_size

size_in_bytes = get_directory_size("/home/biswamitra/health/knee_data/test/multicoil_test")
print(size_in_bytes)
```

```python
import os
import h5py
import numpy as np
dir_path = "/home/biswamitra/health/knee_data/train/multicoil_train"
output_path = "/home/biswamitra/health/knee_data/train/deconstructed_train"

for file_names in os.listdir(dir_path):
    final_path = os.path.join(dir_path,file_names)
    with h5py.File(final_path,'r') as f:
        data = f['kspace']
        slices, coils, x, y = data.shape
        prefix = os.path.splitext(file_names)[0]
        for slice in range(slices):
            curr_slice = data[slice]
            slice_name = f"{prefix}_{slice:03d}.npy"
            pth = os.path.join(output_path,slice_name)
            np.save(pth,curr_slice)
```
