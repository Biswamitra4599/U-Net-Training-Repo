### 🔑 Key Lesson: First/Last Slices Are Often Empty

In 3D MRI volumes:
- **Slice 0 and last slices** often contain NO anatomy, just noise
- **Middle slices** contain the actual anatomical structures
- Always use `volume.shape[0] // 2` to get a representative slice

That's why `fs_coil_size_filtered[100]` worked (it's from middle of dataset) but `hf['kspace'][0]` showed nothing!

## Key Takeaway: Always Normalize MRI Images for Visualization

**The reconstruction works fine - it's just the display!**

Use percentile clipping (1-99% or 2-98%) for best viewing results. This handles outliers from:
- Noise
- Motion artifacts  
- Aliasing from undersampling
- DC offset in k-space center

```python
# Utility function for proper MRI visualization
def normalize_for_display(image, method='percentile', p_low=1, p_high=99):
    """
    Normalize MRI image for proper visualization
    
    Args:
        image: torch.Tensor or np.ndarray
        method: 'percentile' (recommended) or 'minmax'
        p_low, p_high: percentile range for clipping (only for 'percentile')
    
    Returns:
        Normalized image as numpy array [0, 1]
    """
    import torch
    import numpy as np
    
    # Convert to tensor if needed
    if isinstance(image, np.ndarray):
        image = torch.from_numpy(image)
    
    if method == 'percentile':
        vmin = torch.quantile(image, p_low/100)
        vmax = torch.quantile(image, p_high/100)
        image_clipped = torch.clamp(image, vmin, vmax)
        normalized = (image_clipped - vmin) / (vmax - vmin + 1e-8)
    else:  # minmax
        normalized = (image - image.min()) / (image.max() - image.min() + 1e-8)
    
    return normalized.numpy()

# Example usage:
# plt.imshow(normalize_for_display(rss_image_data), cmap='gray')
```
